### (1) 함수와 클래스

#### 함수 정의

```python
def calc_func(a, b): # outer 
    x = a # 10
    y = b # 20
    
    # inner
    def plus():
        p = x + y
        return p
    
    def minus():
        m = x - y
        return m

    return plus, minus

p, m = calc_func(10, 20)
print('plus =', p())
print('minus=', m())
```

#### 위 함수를 클래스로 구현

```python
def calc_func(a, b): # outer 
    x = a # 10
    y = b # 20
    
    # inner
    def plus():
        p = x + y
        return p  
        # return print('p = ', p)
    
    def minus():
        m = x - y
        return m 
        # return print('m = ', m)

    return plus, minus
    # return print('minus = ', m,'\n', 'plus =', m)
    
# calc_func(10, 20)

p, m = calc_func(10, 20)

# # print('p 저장 값 =', p)
# # print('m 저장 값=', m)

print('plus =', p())
print('minus=', m())
```

#### 클래스 정의 및 구성 요소합니다.

```python
# - 클래스 하나로 서로 다른 유닛 3개를 만들었는데, 이렇게 만들어진 유닛들을 객체(object)라고 합니다. 즉, soldier1, soldier2, tank는 객체이고, 붕어빵도 객체입니다. - 그리고 이렇게 만들어진 객체를 클래스의 인스턴스(instance)라고 합니다.
# - 즉, soldier1, soldier2, tank 객체는 Unit 클래스의 인스턴스이고, 붕어빵은 붕어빵 틀의 인스턴스입니다.
  
# -- TIP | 객체와 인스턴스는 사실 같은 개념입니다.
# -- 객체를 만드는 것은 결국 클래스의 인스턴스를 만드는 것입니다.
# -- 보통 객체를 단독으로 부를 때는 객체로, 클래스와 연결지어 부를 때는 인스턴스로 표현합니다.
```

#### 클래스 정의 및 구성 요소
##### 1.변수 선언
##### 2.생성자(__init__) : Self 객체 생성 + [멤버변수 초기화] 
##### 3.소멸자 : 객체 사용 완료시 자동으로 메모리에서 소멸 > 파이썬에서는 메모리 관리를 자동으로 해주기 때문에 많이 사용하지 않는다.
##### 4.메소드(함수) 생성
##### 5. 객체들(인스턴스들) 생성
##### 6. 객체 사용

```python
class calc_class :
    # 변수 선언
    x = y = 0  
    # 생성자 : 객체 생성 + [멤버변수 초기화] 
    def __init__(self, a, b):
        print('~~ self 객체 생성~~')
        self.x = a # 10
        self.y = b # 20   
        
    # 메소드(함수1) 정의
    def plus(self): 
        p = self.x + self.y
        return p  
        
    # 메소드(함수2) 정의    
    def minus(self):
        m = self.x - self.y
        return m  
        
    # 소멸자:생성자 메모리에서 삭제 : 
    def __del__(self):
        del self.x
        del self.y
```

##### obj1 객체(인스턴스) 생성

```python
# print('calc_class(10, 20) = ', calc_class(10, 20))
# class(1) -> object(n) 생성 
obj1 = calc_class(10, 20) # obj1는 객체이면서 calc_class(10, 20)의 인스턴스
# print('obj1 = ', obj1)

# object.member()  
print('plus = ', obj1.plus()) # plus =  30
print('minus =', obj1.minus()) # minus = -10
```

##### obj2 객체(인스턴스) 생성

```python
obj2 = calc_class(100, 50) # 생성자 -> 객체2
# print('obj2:', obj2)

print('plus = ', obj2.plus()) # plus =  150
print('minus =', obj2.minus()) # minus = 50
# print()
```

#### 클래스 다른 예

```python
class Car:
    # 1.멤버 변수
    cc = 0 # 엔진 cc
    door = 0 # 문짝 개수
    carType = None # null

    # 2.생성자
    def __init__(self, cc, door, carType):
        # 멤버 변수 초기화
        self.cc = cc
        self.door = door
        self.carType = carType # 승용차, SUV

    # 3.(멤버) 메소드
    def display(self):
        print("자동차는 %d cc이고, 문짝은 %d개, 타입은 %s 이다."
              % (self.cc, self.door, self.carType))

# 4.object 생성
car1 = Car(2000, 4, "승용차")  # 객체 생성 + 초기화
# car1 = Car(2000, 4)

# 5.object.member()
car1.display()  # info(kind) -> display(kind)
print()

car2 = Car(3000, 5, "SUV")
car2.display()
```
